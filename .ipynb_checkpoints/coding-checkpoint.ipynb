```python
import os, glob
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
npy_paths = glob.glob("data/training_processed/patient*/*.npy")
len(npy_paths)
```

```python
MM = []
mm = []
for p in tqdm(npy_paths):
    npy = np.load(p)
    # npy = np.clip(npy, -200, 2000)
    M = npy.max()
    m = npy.min()
    MM.append(M)
    mm.append(m)
```

```python
H = plt.hist(MM)
```
